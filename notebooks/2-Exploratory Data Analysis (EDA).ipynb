EDA is an approach to analyzing the data using visual techniques. It is used to discover trends, and patterns, or to check assumptions with the help of statistical summaries and graphical representations.

For this dataset we will look at the relationship between the independeent features - how they affect each other

```python
import matplotlib.pyplot as plt

import seaborn as sns

import pandas as pd
```

```python
# read and import dataset
df = pd.read_csv("D:/Medical Insurance Price Prediction/data/raw/insurance.csv")

df
```

```python
# check for null values
df.isnull().sum()
```

```python
# explore relationship between categorical columns

cat_features = ['sex', 'smoker', 'region']

plt.figure(figsize=(18, 6))

for i, col in enumerate(cat_features):
    counts = df[col].value_counts()
    plt.subplot(1, 3, i+1)
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='sex', hue='smoker')
plt.title('Sex vs Smoker Status')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Smoker')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='region', hue='smoker')
plt.title('Region vs Smoker Status')
plt.xlabel('Region')
plt.ylabel('Count')
plt.legend(title='Smoker')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='sex', hue='region')
plt.title('Sex vs Region')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Region')
plt.show()
```

```python
features = ['sex', 'smoker', 'region', 'children'] # These features influence insurance charges

# I want to group each by the insurance charges


plt.subplots(figsize=(20, 10))

for i, col in enumerate(features):
    plt.subplot( 2, 2, i+1)
    group_features = df.groupby(col)['charges'].mean().astype(float)
    group_features.plot.bar()

plt.tight_layout()
plt.show()
```
